```python
# m2, pret
apartamente = [
    (70, 100),
    (70, 150),
    (100, 120),
    (30, 50),
    (50, 70),
    (50, 60),
    (100, 500)
]

x = [pair[0] for pair in apartamente]
y = [pair[1] for pair in apartamente]

def functie_liniara(x):
    return x*1.875 - 6.25

y_prezis = [functie_liniara(i) for i in x]

plt.scatter(x, y)
plt.plot(x, y_prezis)
plt.xlabel("metri patrati")
plt.ylabel("kEUR")
plt.show
```

```python
# m2, pret
apartamente = [
    (70, 100),
    (70, 150),
    (100, 120),
    (30, 50),
    (50, 70),
    (50, 60),
    (100, 500)
]

x = [pair[0] for pair in apartamente]
y = [pair[1] for pair in apartamente]

rezultat = stats.linregress(x, y)
slope, intercept = rezultat.slope, rezultat.intercept

def functie_liniara(x):
    return x * slope + intercept

y_prezis = [functie_liniara(i) for i in x]

plt.scatter(x, y)
plt.plot(x, y_prezis)
plt.xlabel("metri patrati")
plt.ylabel("kEUR")
plt.show
```
