```python
x = [10, 20, 30, 40]
y = [10, 20, 30, 40] # lista de valori reale

def functie_liniara(x):
    # a = 1, b = 0
    return x

model = list(map(functie_liniara, x)) # lista de valori estimate (gasite de catre model)

plt.scatter(x, y, c = "red")
plt.xlabel("Coordonata x")
plt.ylabel("Coordonata y")
plt.plot(x, model)
plt.show()

```

```python
x = [10, 20, 30, 40]
y = [10, 20, 30, 40] # lista de valori reale

def functie_liniara(x):
    # a = 1, b = 0
    return x + 1

model = list(map(functie_liniara, x)) # lista de valori estimate (gasite de catre model)

plt.scatter(x, y, c = "red")
plt.xlabel("Coordonata x")
plt.ylabel("Coordonata y")
plt.plot(x, model)
plt.show()

```
