```python
new_x, new_y = make_regression(n_samples = 20, n_features = 1, n_targets = 1, noise = 10)
plt.scatter(new_x, new_y)
plt.show()
```

```python
new_model = LinearRegression()
new_model.fit(new_x, new_y)
```

```python
pred_y = new_model.predict(new_x)
```

```python
plt.scatter(new_x, new_y, c = "red")
plt.plot(new_x, pred_y)
plt.show()
```

```python
real_v_prezis = {"y_real":new_y, "y_prezis":pred_y}
df_real_vs_prezis = pd.DataFrame(real_v_prezis)
df_real_vs_prezis["delta"] = df_real_vs_prezis["y_real"] - df_real_vs_prezis["y_prezis"]
df_real_vs_prezis
```
