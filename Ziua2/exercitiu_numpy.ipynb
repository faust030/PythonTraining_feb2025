```python
######### Exercitiul ################
# Adunați două matrici de dimensiuni egale (de exemplu, 3x3).
```

```python
arr = np.ones(9, dtype=int).reshape(3,3)
arr_1 = arr
arr_2 = arr*2

print("arr_1 =",arr_1)
print("arr_2 =",arr_2)
print("arr_1 + arr_2 =",arr_1 + arr_2)
```

```python
######### Exercitiul ################
# Înmulțiți două matrici de dimensiuni corespunzătoare (de  exemplu, 2x3 și 3x2).
```

```python
arr_1 = np.ones(6).reshape(2,3)
arr_2 = np.ones(6).reshape(3,2)

print("arr_1 = ", arr_1)
print("arr_2 = ", arr_2)
print("arr_1 * arr_2", np.dot(arr_1, arr_2))
```
